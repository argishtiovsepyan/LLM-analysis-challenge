# Research

`rlhf_reward_model` (https://huggingface.co/OpenAssistant/reward-model-deberta-v3-large-v2)

**Reinforcement Learning Human Feedback Reward Model**

- Uses the transformer architecture to learn the relationships between words and phrases.
- Has 137B parameters (large number of learnable weights) and trained on a dataset of 1.7M question-answer pairs.
- Fine-tuned on a dataset of human feedback, specifically trained to predict human ratings of the quality of answers.
- Used for QA model evaluation, as a reward score in RLHF, and to detect potential toxic responses via ranking.

`reciprocate_reward_model` (https://huggingface.co/reciprocate/gpt-j_rm_format-oa)

**Reciproate Reward Model**

- Uses the transformer architecture to learn the relationships between words and phrases.
- Has 175B parameters and fine-tuned on a 500B dataset of text.
- Used to evaluate the quality of conversational AI assistants, can measure the politeness, informativeness, and relevance of the responses.

The two reward models are used to generate a score for `completions` from the initial `prompt`. The score is a measure of how well the response matches the prompt and how informative it is. The higher the score, the better the response.

|| `rlhf_reward_model` | `reciprocate_reward_model` |
|---|---|---|
| **Type** | Specialized | General Purpose |
| **Evaluation** | Quality of responses to questions | Quality of conversational text |
| **Fine-tuning** | Fine-tuned on human feedback | Not fine-tuned on human feedback |
| **Application** | QA model evaluation | Evaluating the quality of conversational AI assistants |

```python
def get_list_lengths(df):
    """
    Calculate the lengths of lists contained in each column.
    
    Args:
        df (pd.DataFrame): The input DataFrame with potential list columns.
    
    Returns:
        Dict[str, int]: A dictionary where keys are column names and values are the 
                        unique lengths of lists found within that column.
    """
    
    lengths = {}
    
    for col in df.columns:
        list_cols = df[col].apply(lambda x: isinstance(x, list))
        if list_cols.any():
            unique_lengths = df[col][list_cols].apply(len).unique()
            if len(unique_lengths) == 1:
                lengths[col] = unique_lengths[0]
            else:
                lengths[col] = sorted(unique_lengths)
                
    return lengths
```

```python
def combine_columns(row, columns):
    """Helper function combines specified list columns of a row into a list of tuples."""
    
    return list(zip(*[row[col] for col in columns]))
```

```python
def explode_list(df, columns=None):
    """
    Explode list columns in the DataFrame to ensure synchronized alignment.
    
    This function aligns each specified list column such that the nth element of each 
    list in the columns corresponds to the nth element of the other list columns. It 
    ensures that the elements which originally were at the same position in their 
    respective lists end up on the same row.
    
    Args:
        df (pd.DataFrame): The input DataFrame with potential list columns.
        columns (Optional[List[str]]): List of columns to explode. If not provided, 
                                       columns containing lists will be identified and used.
                                       
    Returns:
        pd.DataFrame: DataFrame with exploded list columns aligned.
    """

    if columns is None:
        columns = [c for c in df.columns if df[c].apply(lambda x: isinstance(x, list)).any()]

    df['combined'] = df.apply(combine_columns, axis=1, columns=columns)
    df = df.explode('combined')

    df[columns] = pd.DataFrame(df['combined'].tolist(), index=df.index)
    df = df.drop(columns='combined')
    
    return df
```

```python
def filter_out_unsuccessful_completions(df):
    """
    Filter out rows with unsuccessful completions.

    Args:
        df (pd.DataFrame): Dataframe containing a 'completions' column.

    Returns:
        pd.DataFrame: Filtered DataFrame.
    """
    
    df_filtered = df.dropna(subset=['completions'])
    df_filtered = df_filtered[
        df_filtered['completions'].apply(lambda completions: "" not in completions)
    ]
    
    return df_filtered
```

```python
def plot_histogram(
    df,
    list_column,
    title = None,
    xlabel = None,
    color = 'blue',
    figsize = (10, 6)
):
    """
    Plots a histogram for a given list column using seaborn and matplotlib.
    
    Args:
        df (pd.DataFrame): The input DataFrame with potential list columns.
        list_column (str): List column to be exploded and plotted.
        title (str, optional): Title of the plot. If not provided, a default will be used.
        xlabel (str, optional): Label for the x-axis. If not provided, a default will be used.
        color (str, optional): Color of the histogram. Default is 'blue'.
        figsize (tuple, optional): Size of the figure. Default is (10, 6).
    """
    
    if title is None:
        title = f'Original Distribution of {list_column}'
    
    if xlabel is None:
        xlabel = 'Scores'
    
    fig, ax = plt.subplots(figsize=figsize)
    sns.histplot(df[list_column].explode(), ax=ax, color=color, kde=True)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
def compute_stats(df, list_column):
    """
    Compute mean, standard deviation, count, standard error, and 95% confidence intervals for a specific list column.
    Group by 'uids' column and return the rows sorted by the lower confidence interval.

    Args:
        df (pd.DataFrame): The input DataFrame.
        list_column (str): The column to compute statistics for.

    Returns:
        pd.DataFrame: A DataFrame with the computed statistics.
    """
    
    stats = df.groupby('uids')[list_column].agg(['mean', 'std', 'count'])
    
    stats['se'] = stats['std'] / np.sqrt(stats['count'])
    
    stats['ci_lower'] = stats['mean'] - 1.96 * stats['se']
    stats['ci_upper'] = stats['mean'] + 1.96 * stats['se']
    
    sorted_lower = stats.sort_values(by='ci_lower', ascending=False)
    sorted_mean = stats.sort_values(by='mean', ascending=False)
    
    return sorted_lower, sorted_mean
```

### Strategy

To develop a strategy, the normalized distribution for individual servers was plotted against the overall normalized distribution of all servers. The focus is on servers whose means exceed the upper confidence bounds. This deviation suggests anomalous behavior that is statistically significant, potentially pointing to exploitation of the reward system. These servers can subsequently be flagged for further investigation. This approach mirrors fraud detection techniques used in the financial industry, framing the task as a binary classification. Is the server exploiting the reward system or not?

For an effective detection strategy, a Random Forest can be utilized as the baseline model. Once a collection of flagged servers is accumulated, they can be leveraged as training data to gain insights into feature significance and importance. The model would learn behavioral patterns of flagged servers by evaluating which features play a pivotal role in the classification process. This establishes a foundational understanding of flagged servers and creates a preliminary layer of detection. After this training phase, the model can be applied to the entire dataset, in essence creating a train-test split, to classify and potentially uncover additional servers based on the learned patterns.

To strengthen the detection mechanism, an unsupervised model such as KNN (K-Nearest Neighbors) can be utilized to recognize subtle non-linear relationships and patterns that the Random Forest might overlook. The clusters formed through KNN can be examined to discern analogous behavioral patterns amongst flagged servers. This can potentially reveal shared exploitation strategies of how servers could be manipulating the reward system. Integrating this additional layer ensures not just a reinforced detection system, but also a more profound understanding of server behaviors, offering a comprehensive strategy against exploitation.

Finally, by implementing NLP (Natural Language Processing) on the prompts and completions, it's possible to identify common tokens or phrases. Analyzing these patterns can offer insights into whether specific wordings or topics consistently result in reward model overperformance, allowing for more nuanced tuning of the reward system.

### Recommendations

- Adjust reward distribution thresholds based on flagged server behavior.
- Engineer new features based on observed patterns that indicate exploitation to enhance detection accuracy and reduce the influence of features susceptible to these exploitations.
- Refine detection models with feedback from Random Forest and KNN predictions and periodically retrain to keep up with evolving exploitation techniques.
- Modify the reward model to be less sensitive or penalize behaviors consistently linked to exploitation.
